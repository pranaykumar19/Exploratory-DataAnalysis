```python
from matplotlib import pyplot as plt
players = ['pranay', 'virat', 'dhoni', 'warner', 'jofra']
score = [74, 105, 255, 63,99]
plt.bar(players,score, color = 'Black')
plt.title("stats")
plt.xlabel('players')
plt.yLabel('score')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
cars = 'RENAULT', 'KIA', 'BMW', 'AUDI'
units = [15, 21, 11, 10]
explode = [0.1, 0, 0, 0]
fig1, ax1 = plt.subplots()
ax1.pie(units, explode=explode, labels=cars, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```
